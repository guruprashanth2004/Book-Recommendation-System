Book recomondation collaborative based

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

importing datasets

```python
books=pd.read_csv('/Books.csv',sep=',',on_bad_lines='skip',encoding='latin-1')
```

```python
books.head(2)
```

```python
books.shape
```

```python
books=books[['ISBN','Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher',
       'Image-URL-L']]
```

```python
books.head(2)
```

dataset modifications

```python
books.rename(columns={
    "Book-Title":"Title",
    "Book-Author":"Author",
    "Year-Of-Publication":"year",
    "Publisher":"publisher",
    "Image-URL-L":"image_url"
}, inplace=True)
```

```python
books.head(2)
```

```python
users=pd.read_csv('/Users.csv',sep=',',on_bad_lines='skip',encoding='latin-1')
```

```python
users.head(2)
```

```python
ratings=pd.read_csv('/Ratings.csv',sep=',',on_bad_lines='skip',encoding='latin-1')
```

```python
ratings.head(2)
```

filtering data

```python
#minimizing data
x=ratings['User-ID'].value_counts()>200
```

```python
y=x[x].index
```

```python
ratings=ratings[ratings['User-ID'].isin(y)]
```

```python
rating_books=ratings.merge(books,on='ISBN')
```

```python
rating_books.head(2)
```

```python
num_rating=rating_books.groupby('Title')['Book-Rating'].count().reset_index()
```

```python
num_rating.head(2)
```

```python
num_rating.rename(columns={
    'Book-Rating':'num_book_rating'
},inplace=True)
```

```python
num_rating.head()
```

final modified data set

```python
final_rating_books=rating_books.merge(num_rating, on='Title')
```

```python
final_rating_books=final_rating_books[final_rating_books['num_book_rating']>=50]
```

```python
final_rating_books.head()
```

```python
#removing duplicates
final_rating_books.drop_duplicates(['User-ID','Title'],inplace=True)
```

clustering

```python
#creating pivot table
book_pivot=final_rating_books.pivot_table(columns='User-ID',index='Title',values='Book-Rating')
```

```python
book_pivot
```

```python
#filling null values
book_pivot.fillna(0,inplace=True)
```

```python
#using scipy libary
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

fitting model

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion =model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
#indexing book titles
for i in range(len(suggestion)):
  print(book_pivot.index[suggestion[i]])
```

```python
books_name = book_pivot.index
```

creating pkl files

```python
import pickle
```

```python
with open('model.pkl','wb') as f:
  pickle.dump(model,f)
with open('books_name.pkl','wb') as f1:
  pickle.dump(books_name,f1)
with open('final_rating_books.pkl','wb') as f2:
  pickle.dump(final_rating_books,f2)
with open('book_pivot.pkl','wb') as f3:
  pickle.dump(book_pivot,f3)
```

checking model

```python
def rec_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distance, suggestion =model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
  for i in range(len(suggestion)):
    books=book_pivot.index[suggestion[i]]
    for j in books:
      print(j)
```

```python
book_name= 'Animal Farm'
rec_book(book_name)
```
